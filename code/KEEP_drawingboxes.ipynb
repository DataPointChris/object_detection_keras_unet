## This image

```python
#Threshold from binary image
thresh2 = cv2.imread('../images/network_training/mask/0/austin11.tif')
thresh2_gray = cv2.cvtColor(thresh2,cv2.COLOR_BGR2GRAY) 
```

```python
ret, thresh3 = cv2.threshold(thresh2_gray, 127, 255, 0)
```

```python
#calculate the contours from binary image
contours, hierarchy = cv2.findContours(thresh3, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE) 
```

```python
box1 = np.zeros((thresh2.shape[0], thresh2.shape[1], 3), dtype=np.uint8)
```

```python
for cnt in contours:
    x,y,w,h = cv2.boundingRect(cnt)
    cv2.rectangle(box1,(x,y),(x+w,y+h),(0,255,0),5)
```

```python
fig, ax = plt.subplots(figsize=(24,24), ncols=2)
ax[0].imshow(box1)
ax[1].imshow(image)
```
